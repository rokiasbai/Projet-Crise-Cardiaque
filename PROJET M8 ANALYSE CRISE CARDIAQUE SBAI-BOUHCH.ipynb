# REGRESSION LINEAIRE MULTIPLE AVEC TOUTES LES VARIABLES

```python
#Fonction de la rgression linaire
def ma_reg(X,y):
    '''
    X : numpy array de dimention n x p
    y : numpy array de dimention n
    '''
    import numpy as np
    
    n,p = X.shape
    m = np.prod(y.shape)
    if n != m:
        raise Exception('X doit etre une matrice de n lignes et p colonnes et y un vecteur de n lignes')
        
    a = np.linalg.solve(X.T@X,X.T@y)
    z = X@a
    e = y - z # z - y
    s2 = np.sum(e**2)/(n-p)
    SCT = np.sum((y-np.mean(y))**2)
    SCE = np.sum(e**2)
    R2 = 1 - SCE/SCT
    H = X@np.linalg.solve(X.T@X,X.T)  # plus stable et plus rapide
    h = np.diag(H)
    c = h/((1-h)**2)*(e**2)/(p*s2)
        
    return a, s2, R2,z,e,h,c
```

```python
y = data["output"]
x = data.drop(["output"],axis = 1)
X_reg = np.concatenate((x,np.ones((n,1))), axis=1)
a, s2, R2,z,e,h,c =  ma_reg(X_reg, y)
print(R2) #Qualit de la rgression linaire avec toutes les variables
```

# Regression multiple en ajoutant une  une les variables

```python
n,p = x.shape
```

```python
xreg = x.to_numpy()
```

```python
#Qualit de la rgression linaire en ajoutant une  une les variables
A = np.zeros((2, p))
s2 = np.zeros((p,))
R2 = np.zeros((p,))
sx2 = np.zeros((p,))
for i in range(p):
    X_reg = np.stack([xreg[:,i], np.ones(y.shape)], axis=-1)
    sx2[i] = np.sum((xreg[:,i]-np.mean(xreg[:,i]))**2)  
    A[:,i], s2[i], R2[i],zz,ee,hh,cc =  ma_reg(X_reg, y)
with np.printoptions(precision=3, suppress=True):
    print('Variance estime :',s2) 
    print('Variance calcule :',sx2)
    print('Coeff          R2 :',R2)
    print(' a et b        A  :',A)
```

# Rgression multiple en enlevant les points aberrants 

```python
#Reprage des points aberrants :
Tab = np.array([z,e,h,c,c>(4/n)])
print('-------------------------------------------')
print('     z      e       h       c     Aberrant ')
print('-------------------------------------------')
np.set_printoptions(threshold = np.inf)
with np.printoptions(precision=4,suppress=True):
       print(Tab.T)
```

```python
#Suppression des points aberrants:
Ab = Tab[4]
Td = []
for i in range(n):
    if(Ab[i]==1):
        Td.append(i)
x2 = x.drop(Td)
y2 = y.drop(Td)
n1,p1 = x2.shape
X_reg = np.concatenate((x2,np.ones((n1,1))), axis=1)
a, s2, R2,z,e,h,c =  ma_reg(X_reg, y2)
print(R2)#Qualit de la rgression linaire avec toutes les variables sans les observations aberrantes
```

# Rgression en utilisant les donnes de lACP

```python
X=xreg
Xr = (X - np.mean(X,axis=0))/np.std(X,axis=0)
cov=(Xr.T@Xr)/n
w,v = np.linalg.eigh(cov)
w=w[::-1]
v=v[:,::-1]

plt.bar(np.arange(p), height=(np.cumsum(w)/np.sum(w)) *100)
plt.plot((w/np.sum(w))*100,'r-o')
plt.grid(True)
```

# Test de student sur la pente de rgression

H0 = independance (a=0)

H1 = dependance (a!=0)

```python
A = np.zeros((2, p))
s2 = np.zeros((p,))
R2 = np.zeros((p,))
sx2 = np.zeros((p,))
for i in range(p):
    X_reg = np.stack([xreg[:,i], np.ones(y.shape)], axis=-1)
    sx2[i] = np.sum((xreg[:,i]-np.mean(xreg[:,i]))**2)  
    A[:,i], s2[i], R2[i],z,e,h,c =  ma_reg(X_reg, y)
```

```python
from scipy.stats import t
import scipy.stats as st
ddl=n-2
alpha=0.05
t = np.zeros((p,))
pval = np.zeros((p,))
for i in range(p):
    t[i]=(A[:,i][0])/(np.sqrt(s2[i]/sx2[i]))
    pval[i]=st.t.pdf(t[i],ddl)
    if pval[i]<alpha :
        print("H_o est rejete ")
    else:
        print("H_o est garde ")
```
